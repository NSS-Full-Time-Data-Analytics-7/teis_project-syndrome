```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
import pickle

%matplotlib inline
```

```python

# reading a pickle file
with open('../data/teis_final_data.pickle', 'rb') as file:
    final_data = pickle.load(file)
```

```python
total_re_referrals = (final_data.duplicated(subset = 'Child ID', keep = 'first')).sum()
total_re_referrals
#########################TOTAL RE-REFERRALS 18781 
```

```python
((total_re_referrals/len(final_data))*100).round(decimals =2)
########################RATE OF RE-REFFERAL  16.26%  
#dividing 18781/115523
```

```python
re_referrals = final_data[final_data.duplicated(subset = 'Child ID')]
re_referrals
```

```python
numchildren = re_referrals.groupby('Child ID', as_index=False).count()
numchildren['Child ID']

#######there are 18781 re-referrals from above and when we group by child ID we find that there are 15641 children
#######that have been referred more than once. 
##if we divide 15641 children over the total # of children (96742) we get 16.17% children have been referred more than once.
```

```python
poe_reref= re_referrals.groupby(['POE'])['Child ID'].count()
poe_reref = poe_reref.to_frame().reset_index()
poe_reref = poe_reref.rename(columns = {'Child ID' : "Total Re-referrals"})
poe_reref
##this gives up qty of rereferrals by POE (grand total = 18781)
```

```python
poe_ref = final_data.groupby(['POE'])['Child ID'].count()
poe_ref = poe_ref.to_frame().reset_index()
poe_ref = poe_ref.rename(columns = {'Child ID' : 'Total Referrals'})
poe_ref
##this gives us total qty of referrals by POE (grand total = 115523)
```

```python
poe_rerate = pd.merge(poe_ref, poe_reref, how='left').fillna(0)
poe_rerate['Re-referral Rate'] = (poe_rerate['Total Re-referrals'] / poe_rerate['Total Referrals'])*100
poe_rerate
#here we combine both of the above tables and divide the total referrals into the the total referrals to get the 
#individual rereferral rates per each POE
```

```python
poe_reref.sort_values('Total Re-referrals', ascending = False, inplace = True)
poe_reref.plot(x= 'POE', y = 'Total Re-referrals', kind = 'barh', color = '#0077B6', legend = False)

#plt.title("Re-referrals by POE")
plt.ylabel("Point of Entry")
plt.xlabel("Number of Re-referrals")
plt.savefig('poe_reref')

#In presentation and reflects # of Re-referrals by POE. 
```
