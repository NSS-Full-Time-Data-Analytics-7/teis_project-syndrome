```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
referraldf = pd.read_csv('../data/teis_referral.csv')   
monthlydf = pd.read_csv('../data/teis_monthly.csv')   
```

```python
combo=pd.merge(monthlydf, referraldf,  how = 'left')
combo
```

```python
slicerdf = combo[['Child ID', 'POE', 'Notification/ Referral Date']] 
slicerdf   
```

```python
dups_df = slicerdf.loc[slicerdf.duplicated(subset='Child ID', keep=False)]
dups_df
```

```python
dups_df.sort_values(by = "Child ID")  
```

```python
final = dups_df.groupby('Child ID', as_index=False).sum()  #this shows that 15267 Child ID's are duplicates
final
```
