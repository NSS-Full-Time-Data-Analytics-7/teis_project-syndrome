**What is the rate of re-referral?**

How many children have multiple referrals (by year or by Point of  Entry Office (POE))? 

1. get a count for each Child ID that is a duplicate Answer: count = 15267
2. group by year
3. group by POE

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime

%matplotlib inline
```

```python
monthlydf = pd.read_csv('../data/teis_monthly.csv')   
monthlydf.head()      
```

```python
monthlydf.info()
```

```python
# extract year from Notification/ Referral and put it in new columm Referral Year
monthlydf['Referral Year'] = pd.DatetimeIndex(monthlydf['Notification/ Referral Date']).year
monthlydf
```

monthlydf['Referral Year'] = monthlydf['Notification/ Referral Date'].str[-4:]
monthlydf

```python
monthlydf.info()
```

```python
slicerdf = monthlydf[['Child ID', 'POE', 'Referral Year']] 
slicerdf   
```

```python
dups_df = slicerdf.loc[slicerdf.duplicated(subset='Child ID', keep=False)]
dups_df.sort_values(by = "Child ID")
```

```python
dups_df.groupby('Child ID', as_index=False).count()  #15,267 child ID's are duplicates which equates to re-referral
```

```python
dups_double = dups_df.groupby(['POE', 'Child ID'], as_index = False).count()
dups_370631 = dups_double.loc[dups_double['Child ID']== 370631]
dups_370631

#this is proof of duplicates in different POE's. (also child ID 382142)
```

```python
dups_df_Child = dups_df.groupby('Child ID', as_index=False).count()
dups_df_Child.value_counts('Child ID')
```

```python
len(dups_df_Child)/len(monthlydf) * 100   
```

```python
RY = monthlydf[['Child ID', 'Referral Year']] 
RY 
```

```python
RY1 = RY.loc[slicerdf.duplicated(subset='Child ID', keep=False)]
RY1.sort_values(by = "Child ID")
```

```python
RY1.groupby('Referral Year').count().reset_index()   #33573
```

```python
POE = monthlydf[['Child ID', 'POE']] 
POE 
```

```python
POE1 = POE.loc[slicerdf.duplicated(subset='Child ID', keep=False)]
POE1.sort_values(by = "Child ID")
```

```python
POE1.groupby('POE').count().reset_index() #33573
```
