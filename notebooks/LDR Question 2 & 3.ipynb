```python
TEIS_POE = monthlydf[['POE','DOB','IFSP Count', 'Eligibility Count']].fillna(0)
TEIS_POE
```

```python
monthlydf['Notification/ Referral Date'] = pd.to_datetime(monthlydf['Notification/ Referral Date'])
monthlydf['Notification/ Referral Date']
```

```python
monthlydf['DOB'] = pd.to_datetime(monthlydf['DOB'])
monthlydf['DOB']
```

```python
dob = monthlydf['Notification/ Referral Date'] - monthlydf['DOB']
dob
```

```python
dob.mean()
```

```python
print ('Average age is 467 days')  #total average age of all referrals
```

```python
TEIS_POE['Avg_Age'] = (monthlydf['Notification/ Referral Date'] - monthlydf['DOB'])
TEIS_POE['Avg_Age'] = TEIS_POE['Avg_Age'].dt.days.astype(int)
TEIS_POE
```

```python
Age = TEIS_POE.groupby('POE')['Avg_Age'].agg(['mean', 'median', 'min', 'max']).round(decimals = 0)
Age
#Evlauation of age by POE
```

A LOW PROPORTION OF REFERRALS

```python
TEIS_POE_Success = TEIS_POE[['POE','IFSP Count', 'Eligibility Count']].groupby('POE').sum()
TEIS_POE_Success
```

```python
TEIS_POE_Success['Eligibility Success'] = (TEIS_POE_Success['IFSP Count']/TEIS_POE_Success['Eligibility Count'] *100).round(decimals =2)
TEIS_POE_Success
```

```python
newdf = ((TEIS_POE.groupby('POE')['IFSP Count'].sum()/TEIS_POE.groupby('POE')['IFSP Count'].count())*100).to_frame().sort_values(by = "IFSP Count")
newdf
```

```python
newdf.rename(columns = {'IFSP Count':'IFSP Rate'}, inplace = True)
newdf
```

```python
TEISmerge1=pd.merge(TEIS_POE_Success, newdf, on = ['POE'], how = 'inner')
TEISmerge1
```

```python
TEISALL=pd.merge(TEISmerge1, Age, on = ['POE'], how = 'inner')
TEISALL
```
