```python
import pandas as pd
import operator
import time
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
TEIS_Monthly = pd.read_csv('../teis_project-syndrome/data/TEIS_Monthly.csv')
```

```python
TEIS_Monthly.head(5)
```

```python
TEIS_Monthly['Notification/ Referral Date'] = pd.to_datetime(TEIS_Monthly['Notification/ Referral Date'])
TEIS_Monthly['Notification/ Referral Date']
```

```python
TEIS_Monthly['DOB'] = pd.to_datetime(TEIS_Monthly['DOB'])
TEIS_Monthly['DOB']
```

```python
TEIS_Referral = TEIS_Monthly[['Referral Source Type Name','IFSP Count']].fillna(0)
TEIS_Referral.head(5)
```

```python
TEIS_Referral_Group_Sum = TEIS_Referral.groupby('Referral Source Type Name').sum()
TEIS_Referral_Group_Sum
```

```python
TEIS_Referral_Group_Count = TEIS_Referral.groupby('Referral Source Type Name').count()
TEIS_Referral_Group_Count
```

```python
TEIS_Referral_Group_Sum/TEIS_Referral_Group_Count * 100
```

```python
TEIS_POE = TEIS_Monthly[['POE','Eligibility Count','IFSP Count','DOB', 'Notification/ Referral Date']].fillna(0)
```

```python
TEIS_POE['Avg_Age']= (TEIS_Monthly['Notification/ Referral Date'] - TEIS_Monthly['DOB'])
TEIS_POE['Avg_Age'] = TEIS_POE['Avg_Age'].dt.days.astype(int)
TEIS_POE
```

```python
TEIS_AGE_INFO =TEIS_POE.groupby('POE')['Avg_Age'].agg(['mean', 'median', 'min', 'max']).round(decimals = 0)
TEIS_AGE_INFO
```

```python
TEIS_POE_Success = TEIS_POE[['POE','IFSP Count','Eligibility Count']].groupby('POE').sum()
TEIS_POE_Success
```

```python
TEIS_POE_Success['Eligibility Success'] = (TEIS_POE_Success['IFSP Count']/TEIS_POE_Success['Eligibility Count'] *100).round(decimals =2)
TEIS_POE_Success
```

```python
TEIS_POE_RR = ((TEIS_POE.groupby('POE')['IFSP Count'].sum()/TEIS_POE.groupby('POE')['IFSP Count'].count())*100).to_frame().sort_values(by = 'IFSP Count')
TEIS_POE_RR.rename(columns = {'IFSP Count':'IFSP Rate'}, inplace = True)
TEIS_POE_RR
```

```python
TEIS_Merge1 = pd.merge(TEIS_POE_Success, TEIS_POE_RR, on = ['POE'], how = 'inner')
TEIS_Merge1
```

```python
TEIS_All = pd.merge(TEIS_Merge1, TEIS_AGE_INFO, on = ['POE'], how = 'inner')
TEIS_All
```
